```python
# Importing libraries we will need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt


# Reading the train and test csv files
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Knowing the dimensions of them
print("Train Shape: ",train.shape)
print("Test Shape: ",test.shape)

train_ids = train['Id']
test_ids = test['Id']
```

```python
# Displayinginformation about train file
print(train.info())

# Displayinginformation about test file
print(test.info())
```

```python
# Displaying the number of null values in train file in order to handle them
print(train.isnull().sum())

# Visualizing the number of null values on the heat map
sns.heatmap(train.isnull())
```

```python
# Displaying the number of null values in test file in order to handle them
print(test.isnull().sum())

sns.heatmap(test.isnull())
```

```python
# Storing the column names of categorial values in order fill the missing values using mode (train file)
cat_col_train = ['FireplaceQu','GarageType','GarageFinish','MasVnrType','BsmtQual',
           'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu',
          'GarageQual','GarageCond']

# Storing the column names of numerical values in order fill the missing values using mean (train file)
ncat_col_train = ['LotFrontage','GarageYrBlt','MasVnrArea']

# Storing the column names of categorial values in order fill the missing values using mode (test file)
cat_col_test = ['FireplaceQu','GarageType','GarageFinish','MasVnrType','BsmtQual',
           'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu',
          'GarageQual','GarageCond','MSZoning','Utilities','Exterior1st','Exterior2nd','KitchenQual','Functional','SaleType']

# Storing the column names of categorial values in order fill the missing values using mode (test file)
ncat_col_test = ['LotFrontage','GarageYrBlt','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath',
                'BsmtHalfBath','GarageCars','GarageArea']

# Storing the column names of values that have a lot of missing values which will be hard to fill
to_drop = ['Id','Alley','PoolQC','Fence','MiscFeature']
```

```python
# Filling the missing values of categorial values (train file)
for i in cat_col_train:
    train[i] = train[i].fillna(train[i].mode()[0])
    
# Filling the missing values of numerical values (train file)
for j in ncat_col_train:
    train[j] = train[j].fillna(train[j].mean())

# Filling the missing values of categorial values (test file)
for i in cat_col_test:
    test[i] = test[i].fillna(test[i].mode()[0])
    
# Filling the missing values of numerical values (test file)
for j in ncat_col_test:
    test[j] = test[j].fillna(test[j].mean())

# Dropping them
for k in to_drop:
    train.drop([k], axis = 1, inplace = True)
    test.drop([k], axis = 1, inplace = True)
```

```python
# Displaying the heat map of train file to make sure that there is no missing values
sns.heatmap(train.isnull())
```

```python
# Displaying the heat map of test file to make sure that there is no missing values
sns.heatmap(test.isnull())
```

```python
# Concatenating the test and train files in one data frame
# We did this because there are few categories that are not present but available in test file
# So we will concat them to perfrom one encoding proccess on all categorial columns
final_df = pd.concat([train,test], axis = 0)

# Storing in an array all the column names that we will encode
# ofc we do the ecoding process to only categorial values
# We will do the concat encoding using hot encodeing
all_cat_col = ['MSZoning','Street','LotShape','LandContour','Utilities','LotConfig','LandSlope',
              'Neighborhood','Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl',
              'Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual',
              'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Heating','HeatingQC','CentralAir',
              'Electrical','KitchenQual','Functional','FireplaceQu','GarageType','GarageFinish','GarageQual',
              'GarageCond','PavedDrive','SaleType','SaleCondition']
```

```python
# The Function will use for encoding
def cat_onehot_encoding(multicol):
    df_final = final_df
    i = 0
    for fields in multicol:
        print(fields)
        df1 = pd.get_dummies(final_df[fields],drop_first = True)
        
        final_df.drop([fields], axis = 1, inplace = True)
        if i==0:
            df_final = df1.copy()
        else:
            df_final = pd.concat([df_final,df1], axis=1)
        i = i+1
    
    df_final = pd.concat([final_df,df_final], axis = 1)
    
    return df_final

# Calling the function
final_df = cat_onehot_encoding(all_cat_col)
```

```python
# Removing the duplicate values from the data frame
final_df = final_df.loc[:,~final_df.columns.duplicated()]
final_df.shape
```

```python
# Dividing the data to tain and test data
df_train = final_df.iloc[:1460,:]
df_test = final_df.iloc[1460:,:]

# Displaying the dimesnion after dividing
print("Train Shape: ",df_train.shape)
print("Test Shape: ",df_test.shape)
```

```python
# Putting the feature values in x (ofc removing the SalePrice which is we want to predict)
# Putting the target value in y
x = df_train.drop('SalePrice', axis=1)
y = df_train['SalePrice']

# Spitting our dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
y_pred_linear = linear_model.predict(x_test)
```

```python
# MSE for linear regression
mse_linear = mean_squared_error(y_test, y_pred_linear)
print(f"Mean Squared Error (MSE) for Linear Regression: {mse_linear}")
```

```python
# RMSE for linear regression
rmse_linear = np.sqrt(mse_linear)
print(f"Root Mean Squared Error (RMSE) for Linear Regression: {rmse_linear}")
```

```python
mae_linear = mean_absolute_error(y_test , y_pred_linear)
print(f"Mean Absolute Error (MAE) for Linear Regression: {mae_linear}")
```

```python
r2 = r2_score(y_test , y_pred_linear)
print(f"R-squared: {r2}")
```

```python
x_final_test = df_test.drop('SalePrice', axis=1)

# Make predictions on the final test data
final_predictions = linear_model.predict(x_final_test)

submission = pd.DataFrame({
    'Id': test_ids,  # Use the stored test IDs
    'SalePrice': final_predictions
})

submission.to_csv('house_price_predictions_lr.csv', index=False)
```

```python
# Generate index to match the length of y_test
c = [i for i in range(1, len(y_test) + 1)]

# Plotting the figure
fig = plt.figure()
plt.plot(c, y_test, color="blue", linewidth=2, linestyle="-", label="Actual")
plt.plot(c, y_pred_linear, color="red", linewidth=2, linestyle="-", label="Predicted")

# Adding title, labels, and legend
fig.suptitle('Actual and Predicted', fontsize=20)
plt.ylabel('Life Expecctancy', fontsize=16)
plt.legend()

# Show the plot
plt.show()
```

```python
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(x_train, y_train)
```

```python
y_pred_knn = knn_model.predict(x_test)
```

```python
mse_knn = mean_squared_error(y_test, y_pred_knn)
print(f"Mean Squared Error (MSE) for KNN Regression: {mse_knn}")
```

```python
mae_knn = mean_absolute_error(y_test, y_pred_knn)
print(f"Mean Absolute Error (MAE) for KNN Regression: {mae_knn}")
```

```python
r2_Knn = r2_score(y_test , y_pred_knn)
print(f"R-squared: {r2_Knn}")
```

```python
x_final_test_knn = df_test.drop('SalePrice', axis=1)

# Make predictions on the final test data
final_predictions_knn = knn_model.predict(x_final_test_knn)

submission_minkowski = pd.DataFrame({
    'Id': test_ids,  # Use the stored test IDs
    'SalePrice': final_predictions_knn
})

submission_minkowski.to_csv('house_price_predictions_knn.csv', index=False)
```
