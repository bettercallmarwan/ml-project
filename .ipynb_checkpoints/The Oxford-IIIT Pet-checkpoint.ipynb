```python
# Libraries imported
import os                                                          # For files and folders manipulation (Loading images from folders)
import numpy as np                                                 # For numerical computing
import cv2                                                         # For image processing (reading, resizing and displaying)
import matplotlib.pyplot as plt                                    # For creating charts (confusion matrix and AUC)   
import seaborn as sns                                              # For statistical plots (class distribution as in haetmaps)
from sklearn.model_selection import train_test_split               # For splitting dataset into training and testing sets
from sklearn.preprocessing import LabelEncoder, StandardScaler     # LabelEncoder => Converts categorical labels (like strings) into numerical format.
from sklearn.neighbors import KNeighborsClassifier               
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier               
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
folder_path = "image dataset"  # Path to the dataset
img_size = 64                  # Size to which images will be resized
k = 3                          # Number of neighbors (knn)
```

```python
# A function that :
# 1. Load all images from folders
# 2. Resize the images to a fixed size
# 3. Returns 2 array (images, labels)
def load_images(folder_path, img_size):
    images = []    # To store loaded and resized images
    labels = []    # To store the class names for each image
    
    for label in os.listdir(folder_path): # Iterate over each folder in directory (Each folder represent a class)
                                          # os.listdir(folder_path) lists all folder in the directory
        
        species_path = os.path.join(folder_path, label) # Constructs each folder path
        if os.path.isdir(species_path):  # Ensure it's a folder
            for img_name in os.listdir(species_path): #Iterate over the images inside the class folder
                img_path = os.path.join(species_path, img_name) # Constructs each image path
                img = cv2.imread(img_path) # Read the image
                if img is not None: # Check if the image was successfully loaded
                    img = cv2.resize(img, (img_size, img_size))  # Resize the image
                    # Store the image and its label
                    images.append(img)
                    labels.append(label)
    return np.array(images), np.array(labels) # Return the images and labels


images, labels = load_images(folder_path, img_size)
print(f"Loaded {len(images)} images.") # To ensure that all images are loaded
```

```python
# A function that visualize the distribution of class labels in the dataset

def plot_class_distribution(labels):
    plt.figure(figsize=(10, 7)) # Create a figure
    class_counts = np.unique(labels, return_counts=True) # Count the unique labels
    sorted_indices = np.argsort(class_counts[1])[::-1] # Sort the class labels by their counts
    sorted_labels = class_counts[0][sorted_indices]
    sns.countplot(y=labels, hue=labels, palette="icefire", order=sorted_labels, legend=False) # Plot the class distribution
    plt.title("Number of Images Per Class")
    plt.xlabel("Count")
    plt.ylabel("Class")
    plt.show()


plot_class_distribution(labels)
```

```python
images_flattened = images.reshape(images.shape[0], -1) # Reshaping the images into 1D vector
label_encoder = LabelEncoder() # convert categorical labels into numeric labels
labels_encoded = label_encoder.fit_transform(labels)
```

```python
# To split the dataset into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(
    images_flattened,             # Array for input features, Where each image is flattened to 1D array, This data will be used to train the model
    labels_encoded,               # Array for target labels (encoded to numeric values)
    test_size=0.2,                # 20% of the data will be used for testing and the remaining will be used for training
    random_state=42,              # To ensure the reproducibility of the splitting of the data each time
    stratify=labels_encoded       # To ensure that the train and test datasets contain the same number of images from the same class
)
```

```python
# Scaling to help the model treat all features fairly when making predictions
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# For displaying the confusion matrix
# To visualize how well the model performed by comparing the true labels with the predicted ones using a confusion matrix
def plot_confusion_matrix(y_test, y_pred, label_encoder, metric):
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=label_encoder.classes_, 
                yticklabels=label_encoder.classes_)
    plt.title(f"Confusion Matrix ({metric.capitalize()})")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()


# To display a few images from the test set with the actual labels in form of (true, predected)
def visualize_predictions(X_test, y_test, y_pred, label_encoder):
    plt.figure(figsize=(10, 10))
    for i in range(9):
        img = X_test[i].reshape(img_size, img_size, 3)
        plt.subplot(3, 3, i + 1)
        plt.imshow(img.astype('uint8'))
        plt.title(f"True: {label_encoder.inverse_transform([y_test[i]])[0]}, Pred: {label_encoder.inverse_transform([y_pred[i]])[0]}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()



# To display the ROC curve and calculating the ACU
# This function helps evaluate how well the model is performing by comparing the difference between the true positives and the false positives
def plot_roc_auc(y_test, y_prob, label_encoder):

    # converts the true labels to binary (0, 1)
    y_test_binarized = label_binarize(y_test, classes=np.arange(len(label_encoder.classes_)))
    
    plt.figure(figsize=(10, 8))
    # To calculate the true positives and false positives rate
    for i, class_name in enumerate(label_encoder.classes_):
        fpr, tpr, _ = roc_curve(y_test_binarized[:, i], y_prob[:, i])
        roc_auc = auc(fpr, tpr) # calculats the area under the curve (AUC)
        plt.plot(fpr, tpr, label=f"{class_name} (AUC = {roc_auc:.2f})") # put fpr and tpr on x and y axes
    
    plt.plot([0, 1], [0, 1], 'k--', lw=2)  # Diagonal line for random guessing
    plt.title("ROC Curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend(loc="lower right")
    plt.grid()
    plt.show()

#  Custom Loss Calculation
def calculate_custom_loss(y_true, y_pred):

    misclassifications = np.sum(y_true != y_pred)
    total_samples = len(y_true)
    loss = misclassifications / total_samples
    return loss

from sklearn.metrics import log_loss
def calculate_logistic_loss(y_true, y_pred_proba):

    loss = log_loss(y_true, y_pred_proba)
    return loss
```

```python
logistic_model = LogisticRegression(max_iter=1000, random_state=0)
logistic_model.fit(X_train, y_train)
```

```python
y_pred = logistic_model.predict(X_test)
y_prob = logistic_model.predict_proba(X_test) 
```

```python
# Compute and display accuracy
accuracy = accuracy_score(y_test, y_pred)
loss = calculate_logistic_loss(y_test, y_prob)

print(f"\n{'-' * 50}")
print("Evaluating Logistic Regression")
print(f"{'-' * 50}")
print(f"Accuracy: {accuracy:.2f}")
print(f"Loss: {loss:.4f}")

# Display classification report
print(f"\nClassification Report for Logistic Regression:")
print("-" * 50)
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Plot and display the confusion matrix
plot_confusion_matrix(y_test, y_pred, label_encoder, "Logistic Regression")

# Plot ROC and AUC
plot_roc_auc(y_test, y_prob, label_encoder)

# Visualize predictions
visualize_predictions(X_test, y_test, y_pred, label_encoder)
```

```python
metric = 'manhattan'

knn = KNeighborsClassifier(n_neighbors=k, metric=metric)
knn.fit(X_train, y_train) # Train the knn mode
```

```python
y_pred = knn.predict(X_test) # Predict for teh x_test
y_prob = knn.predict_proba(X_test) # To get the predicted probabilities of each class for each sample in x_test
```

```python
# Compute and display accuracy
accuracy = accuracy_score(y_test, y_pred)
custom_loss = calculate_custom_loss(y_test, y_pred)


print(f"Accuracy: {accuracy:.2f}")
print(f"Loss : {custom_loss:.2f}")

# Display classification report
print(f"\nClassification Report for {metric.capitalize()} Distance:")
print("-" * 50)
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Plot and display the confusion matrix
plot_confusion_matrix(y_test, y_pred, label_encoder, metric)

# Plot ROC and AUC
plot_roc_auc(y_test, y_prob, label_encoder)

# Visualize predictions
visualize_predictions(X_test, y_test, y_pred, label_encoder)
```
